```python
# What is Web Scraping?

# Every webpage is made of HTML and CSS
# To scrape web means to programmatically find specific HTML or CSS content from a webpage.
```

```python
# Load the necessary libraries
import requests
from bs4 import BeautifulSoup as bs
```

```python
# Load the webpage content
r = requests.get('https://quotes.toscrape.com/')

# convert to beautifulsoup object
soup = bs(r.content)

# print the html
print(soup.prettify())
```

```python
# find and find_all
header = soup.find("h1")
print(header)

headers = soup.find_all('h2')
print(headers)
```

```python
# can pass list of elements to look for
first_header = soup.find(['h1', 'h2'])
first_header
```

```python
# pass attributes to the find/find_all function
paragraph = soup.find_all('span', attrs={"class": "text"})
paragraph
```

```python
# Nest find/find_all calls
body = soup.find('body')
div = body.find_all('div')
div
```

```python
# Search for specific strings in our find/find_all calls
import re

paragraphs = soup.find_all('span', string=re.compile('is'))
paragraphs
```

```python
soup.body.prettify()
```

```python
# Select method for selecting CSS
content = soup.select('div span')
content
```

```python
link = soup.select('span.tag-item > a.tag')
link
```

```python
# Grab element by specific property
someelement = soup.select('[aria-hidden=true]')
someelement
```

```python
# Get Different properties of HTML
small = soup.find("small")
print(small.string)

# If multiple child elements us get_text
span = soup.find("span")
print(span.prettify())
print(span.get_text())
```

```python
# Get a specific property from an element
link = soup.find_all('a')

for l in link:
    print(l['href'])
```

```python
# Parents, sibling, child
soup.body.find('div').find_next_siblings()
```

```python
soup.body.find('a').find_parents()
```

```python
soup.find('div').findChildren('h2')
```
